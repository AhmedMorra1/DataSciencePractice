## Exploring Automobile Data Set

I will explore the Automobile Dataset in this notebook.
You can find more about the dataset [here](https://archive.ics.uci.edu/ml/datasets/Automobile)

```python
#Let's start by importing the libraries needed 
import pandas as pd
```

```python
#url of the csv of the data
url = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv"

#let's put the data in a dataframe
df = pd.read_csv(url)

#let's check the dataframe
df.head()
```

```python
df.tail()
```

```python
#The data frame uses the first row as headers let's change that
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df.columns = headers
df.head()
```

```python
#let's get the shape of our data frame
df.shape
```

```python
#now let's drop the missing values in the data frame in the price column 
df.dropna(subset=["price"], axis=0)
df.shape
```

```python
#let's get rid of unwanted columns
#first we see all the columns
pd.set_option('display.max_columns',None)
df.head()
```

```python
#Now let's drop unwanted columns
df.drop(['symboling','normalized-losses'],axis=1,inplace=True)
df.head()
```

```python
#let's save the cleaned dataframe to CSV
df.to_csv("AutomobileDataSet.csv",index=False) # avoid Python/Pandas creating an index in a saved csv
```
