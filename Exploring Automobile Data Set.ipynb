## Exploring Automobile Data Set

I will explore the Automobile Dataset in this notebook.
You can find more about the dataset [here](https://archive.ics.uci.edu/ml/datasets/Automobile)

```python
#Let's start by importing the libraries needed 
import pandas as pd
import numpy as np
```

```python
#url of the csv of the data
url = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv"

#let's put the data in a dataframe
df = pd.read_csv(url)

#let's check the dataframe
df.head()
```

```python
df.tail()
```

```python
#The data frame uses the first row as headers let's change that
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df.columns = headers
df.head()
```

```python
#let's get the shape of our data frame
df.shape
```

```python
#now let's drop the missing values in the data frame in the price column 
df.dropna(subset=["price"], axis=0)
df.shape
```

```python
#let's get rid of unwanted columns
#first we see all the columns
pd.set_option('display.max_columns',None)
df.head()
```

```python
#now let's deal with the missing data
#first we replace ? with np.nan (NaN)
df.replace('?',np.nan,inplace=True)
df.head()
```

```python
#Now let's evaluate missing data
missing = df.isnull()
missing.head()
```

```python
#now let's count missing data
for i in missing.columns.values.tolist():
    print(i)
    print(missing[i].value_counts())
```

```python
#Now we have an idea what is missing let's try to solve it
#let's change missing values in numerical columns with the mean value of the column
df['normalized-losses'].replace(np.nan,df['normalized-losses'].astype('float').mean(axis=0),inplace=True)
df['bore'].replace(np.nan,df['bore'].mean(axis=0),inplace=True)
df['stroke'].replace(np.nan,df['bore'].mean(axis=0),inplace=True)
df['horsepower'].replace(np.nan,df['horsepower'].mean(axis=0),inplace=True)
df['peak-rpm'].replace(np.nan,df['peak-rpm'].mean(axis=0),inplace=True)
df.head()
```

```python
#let's fix the missing data in number of doors by using the mode
print(df['num-of-doors'].value_counts())
print('max value is',df['num-of-doors'].value_counts().idxmax())
```

```python
df['num-of-doors'].replace(np.nan,'four',inplace=True)
df.head()
```

```python
#now let's drop all the 4 rows with no price data
df.dropna(subset=['price'],axis=0,inplace=True)
df.shape
```

```python
#now let's check columns types
df.dtypes
```

```python
#some types are wrong, let's change the wrong types to the right ones
df[['bore','stroke']] = df[['bore','stroke']].astype('float')
df[['horsepower','peak-rpm',"price"]] = df[['horsepower','peak-rpm',"price"]].astype('float')
df[['normalized-losses']] = df[['normalized-losses']].astype('int')
df.dtypes
```

```python
#Now let's do some data standerization
#let's convert these american units to the metric system
df['city-L/100km'] = 235/df["city-mpg"]
df['highway-L/100km'] = 235/df['highway-mpg']
df.head()
```

```python
#now let's do some data normalization to make put different variables values in the same range
df['length'] = df['length'] /df['length'].max()
df['height'] = df['height'] /df['height'].max()
df['width'] = df['width'] /df['width'].max()
df.head()
```

```python
#Let's visualize the horsepower column
%matplotlib inline
import matplotlib.pyplot as plt
df['horsepower'].hist()
```

```python
#Now let's use Binning to divide the hourspower column to high medium low
bins = np.linspace(min(df['horsepower']),max(df['horsepower']),4)
group_name = ['Low','Medium','High']
df['easy hourspower'] = pd.cut(df['horsepower'],bins,labels=group_name,include_lowest=True)
df[['horsepower','easy hourspower']].head()
```

```python
plt.bar(group_name, df["easy hourspower"].value_counts())
```
